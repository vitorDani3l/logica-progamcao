#Tabela Verdade

###And

```python
print(10 > 2 and 4 < 6)   #V or V - V
print(10 < 2 and 4 < 6)   #F or V - F
print(10 > 2 and 4 > 6)   #V or F - F
print(10 < 2 and 4 > 6)   #F or F - F
```

###Or

```python
print(10 > 2 or 4 < 6)   #V or V - V
print(10 < 2 or 4 < 6)   #V or F - V
print(10 > 2 or 4 > 6)   #F or V - V
print(10 < 2 or 4 > 6)   #F or F - F
```

###Not

```python
not 10 > 2   #V - F
```

```python
not 10 < 2   #F - V
```
